```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("tmdb_5000_movies.csv", index_col=3)
df = df[['title', 'budget', 'genres', 'homepage', 'keywords', 'original_language', 'original_title', 'overview', 'popularity', 'production_companies', 'production_countries', 'release_date', 'revenue', 'runtime', 'spoken_languages', 'status', 'tagline', 'vote_average', 'vote_count']]

```

```python
df = df[df["budget"] != 0]
df = df[df["revenue"] != 0]
df = df[df["budget"] > 10000]
df.shape
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.head(20)
```

```python
df["profit"] = df["revenue"] - df["budget"]
df["profit_margin"] = (df["profit"] / df["budget"]) * 100
df["profit_margin"] = df["profit_margin"].round(0).clip(lower=-100, upper=2000).astype(int)
```

```python
df.info()
```

```python
df.tail(5)
```

```python
df = df[df["production_companies"] != "[]"]
df = df[df["production_countries"] != "[]"]

df.tail()
```

```python
df.shape
```

```python
df["release_year"] = df["release_date"].str[:4]
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(df["budget"], df["revenue"], alpha=0.5)
plt.title("Budget vs Revenue")
plt.xlabel("Budget")
plt.ylabel("Revenue")
print(f"Correlation between budget and revenue: {df["budget"].corr(df["revenue"]).round(2)}")
plt.show()
```

```python
df.groupby("release_year")["runtime"].mean().plot()
plt.title("Average Movie Runtime Over the Years")
plt.xlabel("Year")
plt.ylabel("Runtime (minutes)")
plt.show()
```

```python
plt.scatter(df["vote_average"], df["revenue"])
plt.title("Rating vs Revenue")
plt.xlabel("Rating")
plt.ylabel("Revenue")
print(f"Correlation between rating and revenue: {df["vote_average"].corr(df["revenue"]).round(2)}")
plt.show()
```

```python
df["release_month"] = df["release_date"].str[5:7].astype(int)
june = df[df["release_month"] == 6]["profit"].mean().round(0).astype(int)
df.groupby("release_month")["profit"].mean().plot()
plt.title("Profit by Month")
plt.xlabel("Release month")
plt.ylabel("Profit")
print(f"June is the optimal month for releasing as the profit on average was {june}$")
plt.show()
```

```python
def success(id):
    if id["revenue"] > id["budget"] * 3 and id["vote_average"] > 7:
        return "Blockbuster"
    elif id["profit_margin"] > 500:
        return "Surprise Hit"
    elif id["revenue"] < id["budget"]:
        return "Miss"
    else:
        return "Average"


df["verdict"] = df.apply(success, axis=1)
blockbusters = df[df["verdict"] == "Blockbuster"]
misses = df[df["verdict"] == "Miss"]
comp = pd.DataFrame({
    "Average Runtime": [blockbusters["runtime"].mean(), misses["runtime"].mean()],
    "Average Release Month": [blockbusters["release_month"].mean().round(0), misses["release_month"].mean().round(0)],
    "Average Budget": [blockbusters["budget"].mean(), misses["budget"].mean()]
}, index=["Blockbuster", "Miss"])
comp["Average Runtime"].plot(kind="bar", title="Average Runtime", color=["r"])
plt.ylabel("Minutes")
plt.show()

comp["Average Release Month"].plot(kind="bar", title="Average Release Month",color=["g"])
plt.ylabel("Month")
plt.show()

comp["Average Budget"].plot(kind="bar", title="Average Budget")
plt.ylabel("USD")
plt.show()
comp
```
